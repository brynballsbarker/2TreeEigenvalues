```python
import numpy as np
from scipy.sparse import diags
from scipy.linalg import eigh
from math import ceil
```

```python
def compute(n, k):
    A = diags([1, 1], [-1, 1], shape=(n,n)).todense()
    D = np.diag([np.sum(i) for i in A])
    L = D - A
    
    eval_ = eigh(L, eigvals_only=True, eigvals=(k,k))[0]
    
    temp = n**2 *(np.pi**4/12 - n**2 * (np.pi**2 - n**2 * eval_))
    print(temp)
```

```python
for i in range(2,10):
    n = 2**i
    print('n: {}'.format(n))
    
    compute(n, 1)
```

# $n^2\left(\frac{1}{12}\pi^4 - n^2(\pi^2-n^2\lambda_2)\right) \to \frac{1}{360}\pi^6$

```python
def compute(n, k):
    A = diags([1, 1], [-1, 1], shape=(n,n)).todense()
    D = np.diag([np.sum(i) for i in A])
    L = D - A
    
    eval_ = eigh(L, eigvals_only=True, eigvals=(k,k))[0]
    
    temp = n**2 *(4*np.pi**4/3 - n**2 * (4*np.pi**2 - n**2 * eval_))
    print(temp)
```

# $n^2\left(\frac{4}{3}\pi^4 - n^2(4\pi^2-n^2\lambda_3)\right) \to \frac{8}{45}\pi^6$

```python
def compute(n, k):
    A = diags([1, 1], [-1, 1], shape=(n,n)).todense()
    D = np.diag([np.sum(i) for i in A])
    L = D - A
    
    eval_ = eigh(L, eigvals_only=True, eigvals=(k,k))[0]
    
    temp = n**2 * (np.pi**3/3 - n**2 * (2*np.pi - n * (2 - eval_)))
    print(temp)
```

```python
for i in range(2,10):
    n = 2**i
    print('n: {}'.format(n))
    
    compute(n, int(n/2 - 1))
```

# $n^2\left(\frac{1}{3}\pi^3 - n^2\left(2\pi-n(2-\lambda_{n/2})\right)\right) \to \frac{1}{60}\pi^5$
